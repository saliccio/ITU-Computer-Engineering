# Part 1

```python
import numpy as np
import matplotlib.pyplot as plt

def u(n):  # Unit step function
    if n >= 0:
        return 1
    return 0

x1 = lambda n: (n + 2*np.exp(3*n))*u(n-2)  # Signal x1 as a function
x1 = np.vectorize(x1)  # Vectorize the function so that arrays can be given as input to it

def x2(n):  # Signal x2 as a function
    if n < -3:
        return n ** 2
    return 0
x2 = np.vectorize(x2)  # Vectorize the function so that arrays can be given as input to it

def x3(n):  # Signal x3 as a function
    val = 0
    for i in range(n - 2, n + 3):
        val += i * np.exp(i)
    return val
x3 = np.vectorize(x3)  # Vectorize the function so that arrays can be given as input to it

# Get the values for the ranges
x1_values = x1(np.arange(-4, 5))
x2_values = x2(np.arange(-5, 6))
x3_values = x3(np.arange(-3, 4))
```

```python
x1_T = 9  # Period of x1
x1_ak = np.zeros(x1_T, dtype=np.cdouble)    #x1 DTFS coefficients
for k in range(0, x1_T):  # Find the coefficients using the loops
    for n in range(-4, 5):
        x1_ak[k] += x1_values[n]*np.exp(-1j*2*np.pi*k*n/x1_T)
    x1_ak[k] /= x1_T
print("Discrete time Fourier series coefficients for x1: ", x1_ak)

x2_T = 11  # Period of x2
x2_ak = np.zeros(x2_T, dtype=np.cdouble)    #x2 DTFS coefficients
for k in range(0, x2_T):
    for n in range(-5, 6):
        x2_ak[k] += x2_values[n]*np.exp(-1j*2*np.pi*k*n/x2_T)
    x2_ak[k] /= x2_T
print("\nDiscrete time Fourier series coefficients for x2: ", x2_ak)

x3_T = 7  # Period of x3
x3_ak = np.zeros(x3_T, dtype=np.cdouble)    #x3 DTFS coefficients
for k in range(0, x3_T):
    for n in range(-3, 4):
        x3_ak[k] += x3_values[n]*np.exp(-1j*2*np.pi*k*n/x3_T)
    x3_ak[k] /= x3_T
print("\nDiscrete time Fourier series coefficients for x3: ", x3_ak)
```

# Part 2

```python
reobtained_x1 = np.zeros(x1_T, dtype=np.double)
for n in range(0, x1_T):  # Reobtain x1 using the synthesis equation
    for k in range(0, x1_T):
        reobtained_x1[n] += x1_ak[k] * np.exp(1j * k * 2 * np.pi * n / x1_T)
plt.stem(np.arange(-4, 5), reobtained_x1)
plt.title("Signal 1 Reobtained")
plt.show()

reobtained_x2 = np.zeros(x2_T, dtype=np.double)
for n in range(0, x2_T):  # Reobtain x2 using the synthesis equation
    for k in range(0, x2_T):
        reobtained_x2[n] += x2_ak[k] * np.exp(1j * k * 2 * np.pi * n / x2_T)
plt.stem(np.arange(-5, 6), reobtained_x2)
plt.title("Signal 2 Reobtained")
plt.show()

reobtained_x3 = np.zeros(x3_T, dtype=np.double)
for n in range(0, x3_T):  # Reobtain x3 using the synthesis equation
    for k in range(0, x3_T):
        reobtained_x3[n] += x3_ak[k] * np.exp(1j * k * 2 * np.pi * n / x3_T)
plt.stem(np.arange(-3, 4), reobtained_x3)
plt.title("Signal 3 Reobtained")
plt.show()
```

# Part 3

```python
# Non-zero Coefficients: a_-3 = 2,    a_-2 = j,    a_2 = -j,    a_3 = 2
# w0 = pi

def part3_signal_x(t):  # Reobtain the signal
    x = 2 * np.exp(1j * (-3) * np.pi * t) + 1j * np.exp(1j * (-2) * np.pi * t) + (-1j) * np.exp(1j * (2) * np.pi * t) + 2 * np.exp(1j * (3) * np.pi * t)
    return x
part3_signal_x = np.vectorize(part3_signal_x)

plt.figure(figsize=(15, 5))
plt.plot(np.arange(-5, 6, 0.001), part3_signal_x(np.arange(-5, 6, 0.001)))
plt.title("x(t) Reobtained")
plt.show()
```
