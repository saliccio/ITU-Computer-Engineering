```python
import matplotlib.pyplot as plt
import numpy as np

n = np.arange(-10,10)
t = np.linspace(-100, 100, 200)

u_ = lambda n : int(n >= 0)
u = np.vectorize(u_)

x1 = lambda t: t * (u(2 * t) - u(2 * t - 2))
x1 = np.vectorize(x1)

y1 = lambda t: t * t * (u(t) - u(t - 1))
y1 = np.vectorize(y1)

plt.plot(np.convolve(x1(t), y1(t)))
```

```python
import math

x1 = lambda t: (u(t-3) - u(t-5)) * math.sin((2 * math.pi * t) / 3)
x1 = np.vectorize(x1)

def y1(t):
    if t >= -5 and t <= 5:
        return abs(t)
    return 0
y1 = np.vectorize(y1)
plt.plot(np.convolve(x1(t), y1(t)))
```

```python
x1 = lambda t: 2 * math.cos(math.pi * t) * (u(t + 2) - u(t - 2)) - 2 * math.cos(math.pi * (t - 1)) * (u(t + 1) - u(t - 3))
x1 = np.vectorize(x1)

plt.plot(n, x1(n))
```
